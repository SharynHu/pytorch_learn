```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def __getattr__(self, name):
        if name=="course_name":
            return "Math"
        else:
            raise AttributeError("Attribute {} is not found.".format(name))
            
student = Student("Bob", 19)
print(student.__dict__)
# {'name': 'Bob', 'age': 19}
print(student.name)
# Bob
print(student.age)
# 19
print(student.course_name)
# Math
print(student.hobbies)
# AttributeError
print(student.__dict__)
# {'name': 'Bob', 'age': 19}
print(hasattr(student, "course_name"))
# True
print(hasattr(student, "hobbies"))
# False
```

```python
def foo():
    n = 3
    x = 1
    while(x<n):
        yield x
        x += 1
print(foo)
foo = foo()
print(foo)
```
